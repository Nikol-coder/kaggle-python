```python
import pandas as pd
```

```python
df_train=pd.read_csv('./breast-cancer-train.csv')
df_test=pd.read_csv('./breast-cancer-test.csv')
```

```python
df_test_negative=df_test.loc[df_test['Type']==0][['Clump Thickness','Cell Size']]
df_test_positive=df_test.loc[df_test['Type']==1][['Clump Thickness','Cell Size']]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df_test_negative['Clump Thickness'],df_test_negative['Cell Size'],marker='o',s=200,c='red') #负的 红
plt.scatter(df_test_positive['Clump Thickness'],df_test_positive['Cell Size'],marker='x',s=150,c='black') #正的 黑

plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
plt.show

import numpy as np
intercept=np.random.random([1]) #1个数
coef=np.random.random([2]) #两个数
lx=np.arange(0,12)
ly=(-intercept-lx*coef[0])/coef[1]
plt.plot(lx,ly,c='yellow') #随机画一条直线
```

```python
plt.scatter(df_test_negative['Clump Thickness'],df_test_negative['Cell Size'],marker='o',s=200,c='red') #负的 红
plt.scatter(df_test_positive['Clump Thickness'],df_test_positive['Cell Size'],marker='x',s=150,c='black') #正的 黑

plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
plt.show()

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()

#使用前十条数据
lr.fit(df_train[['Clump Thickness','Cell Size']][:10],df_train['Type'][:10])

print('Testing accurary:',lr.score(df_train[['Clump Thickness','Cell Size']],df_train['Type']))
```

```python
#只有10组数据
intercept=lr.intercept_
coef=lr.coef_[0,:]
ly=(-intercept-lx*coef[0])/coef[1]
plt.plot(lx,ly,c='green')

plt.scatter(df_test_negative['Clump Thickness'],df_test_negative['Cell Size'],marker='o',s=200,c='red') #负的 红
plt.scatter(df_test_positive['Clump Thickness'],df_test_positive['Cell Size'],marker='x',s=150,c='black') #正的 黑

plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
plt.show()
```

```python
lr=LogisticRegression()
lr.fit(df_train[['Clump Thickness','Cell Size']],df_train['Type'])
print('Testing accurary:',lr.score(df_train[['Clump Thickness','Cell Size']],df_train['Type']))
```

```python
#运用所有数据
intercept=lr.intercept_
coef=lr.coef_[0,:]
ly=(-intercept-lx*coef[0])/coef[1]
plt.plot(lx,ly,c='blue')

plt.scatter(df_test_negative['Clump Thickness'],df_test_negative['Cell Size'],marker='o',s=200,c='red') #负的 红
plt.scatter(df_test_positive['Clump Thickness'],df_test_positive['Cell Size'],marker='x',s=150,c='black') #正的 黑

plt.xlabel('Clump Thickness')
plt.ylabel('Cell Size')
plt.show()
```
